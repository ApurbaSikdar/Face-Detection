```python
import dlib    ## importing dlib library
```

```python
image = cv2.imread('D:\\Acmegrade\\#Project\\2. Face Detection\\Data Set\\Apurba1.jpg')
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Hog & dlib classifier :
face_detector_hog = dlib.get_frontal_face_detector()
detections = face_detector_hog(image, 1) # 1 – is the scale factor 

```

```python
display (len(detections))
display (detections)
```

```python

for face in detections:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 255), 2)
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()


```

```python
# CNN & dlib Classifier :

cnn_detector = dlib.cnn_face_detection_model_v1('D:\\Acmegrade\\#Project\\2. Face Detection\\Weights\\mmod_human_face_detector.dat')
cnn_detector
```

```python

detections = cnn_detector(image,1)
for face in detections:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  print(c)
  cv2.rectangle(image, (l, t), (r, b), (255, 255, 0), 2)
cv2.imshow('',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
